```python
spider_types = ("black_widow_spider", "funnel_web_spider", "brown_recluse_spider",
                "brazilian_Wandering_spider", "yellow_sac_spider", "brown_widow_spider",
                "american_house_spider", "wolf_spider", "daddy_longlegs_spider",
                "orb_weaver_spider", "jumping_spider", "grass_spider")
```

```python
data_path = Path("/notebooks/spiders")
```

```python
if not data_path.exists():
    data_path.mkdir()
    for s in spider_types:
      dest = (data_path/s)
      dest.mkdir(exist_ok=True)
      search_term = " ".join(s.split("_"))
      results = search_images_bing(key, search_term) 
      download_images(dest, urls=results.attrgot("content_url")) 
```

```python
fns = get_image_files(data_path)
fns
```

```python
failed = verify_images(fns)
failed
```

```python
failed.map(Path.unlink)
```

```python
spiders = DataBlock(
    blocks = (ImageBlock, CategoryBlock),
    get_items = get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed = 0),
    get_y = parent_label,
    item_tfms = Resize(128))
```

```python
dls = spiders.dataloaders(data_path)
```

```python
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
spiders = spiders.new(item_tfms = RandomResizedCrop(224, min_scale=0.6),
                     )
dls = spiders.dataloaders(data_path)
```

```python
learn = cnn_learner(dls, resnet18, metrics=error_rate)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
cm = interp.confusion_matrix()
cm
```

```python
correct = np.diagonal(cm)
```

```python
print("Total : {}".format(cm.sum()))
print("True pos+neg : {}".format(correct.sum()))
print("False pos+neg : {}".format(cm.sum() - correct.sum()))
print("Accuracy : {}".format(correct.sum()/cm.sum() ))
```

```python
to_del = set()
```

```python
cleaner = ImageClassifierCleaner(learn)
cleaner
```

```python
for idx in cleaner.delete():
    to_del.add(cleaner.fns[idx])
len(to_del)
```

```python
for d in to_del:
    d.unlink()
```

```python
spiders_cl = spiders.new(item_tfms = RandomResizedCrop(224, min_scale=0.6),
                     )
dls_cl = spiders_cl.dataloaders(data_path)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
cm = interp.confusion_matrix()
cm
```

```python
correct = np.diagonal(cm)
```

```python
print("Total : {}".format(cm.sum()))
print("True pos+neg : {}".format(correct.sum()))
print("False pos+neg : {}".format(cm.sum() - correct.sum()))
print("Accuracy : {}".format(correct.sum()/cm.sum() ))
```
